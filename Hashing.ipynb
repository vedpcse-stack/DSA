### Leetcode : 128. Longest Consecutive Sequence

```python
nums.sort()
max_seq = 1
if not nums :
    print(0)

prev_ele = nums[0]
temp = 1
for i in nums[1:] :
    if i == prev_ele :
        continue
    if i == prev_ele + 1 :
        temp += 1
        prev_ele = i
        continue
    
    max_seq = max(max_seq , temp)
    temp = 1
    prev_ele = i

max(max_seq , temp)
```

### Leetcode : 560. Subarray Sum Equals K

```python
freq = {0: 1}
prefix_sum = 0
count = 0

for num in nums:
    prefix_sum += num

    if prefix_sum - k in freq:
        count += freq[prefix_sum - k]

    freq[prefix_sum] = freq.get(prefix_sum, 0) + 1

count
```

### Leetcode : 169. Majority Element

```python
d = defaultdict(int)
for i in nums :
    d[i] = d[i] + 1

max_ele = 0
max_count = 0

for i,j in d.items() :
    if j > max_count and j >= len(nums) / 2 :
        max_ele = i
        max_count = j

max_ele
```

### Leetcode : 202. Happy Number

```python
while True :
    d = 0
    sum = 0
    while n > 0 :
        d = n % 10
        sum += d ** 2
        n = n // 10

    if sum == 1 :
        True
    if sum == 4 :
        False
    n = sum
```
