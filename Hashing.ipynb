### Leetcode : 128. Longest Consecutive Sequence

```python
nums.sort()
max_seq = 1
if not nums :
    print(0)

prev_ele = nums[0]
temp = 1
for i in nums[1:] :
    if i == prev_ele :
        continue
    if i == prev_ele + 1 :
        temp += 1
        prev_ele = i
        continue
    
    max_seq = max(max_seq , temp)
    temp = 1
    prev_ele = i

max(max_seq , temp)
```
