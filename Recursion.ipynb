### Leetcode : 39. Combination Sum

```python

def findSum( candidates , target , i , tempArr , ansArr ) :
    
    if sum(tempArr) == target :
        # print(tempArr)
        ansArr.append(tempArr.copy())
        return
    
    if sum(tempArr) > target or i >= len(candidates) :
        return
    
    tempArr.append(candidates[i])
    findSum(candidates , target , i , tempArr , ansArr )
    findSum(candidates , target , i+1 , tempArr , ansArr )

    tempArr.pop()
    findSum(candidates , target , i+1 , tempArr , ansArr )

ansArr = []
target = 8
candidates = [2,3,5]
findSum(candidates , target , 0 , [] , ansArr)

final_ans = []

for i in ansArr :
    if final_ans.count(i) == 0 :
        final_ans.append(i)

final_ans
```

### Leetcode : 40. Combination Sum II

```python

def findSum(arr , target , i , tempArr , ansArr ) :
    if sum(tempArr) == target :
        print(tempArr)
        ansArr.append(tempArr.copy())
        return
    
    if sum(tempArr) > target or i >= len(arr)  :
        return
    
    tempArr.append(arr[i])
    findSum(arr , target , i+1 , tempArr , ansArr)

    tempArr.pop()

    # skip duplicates
    while i + 1 < len(arr) and arr[i] == arr[i+1]:
        i += 1

    findSum(arr , target , i+1 , tempArr , ansArr)

arr = [10,1,2,7,6,1,5]
target = 8
ansArr = []
arr.sort()

findSum(arr , target , 0 , [] , ansArr)

ansArr
```

### Leetcode : 51. N-Queens

```python

def isSafe( row, col, tempArr):
    for r in range(row):
        c = tempArr[r]

        # same column
        # if len(set(tempArr)) == row + 1:
        #     return False
        if c == col :
            return False 

        # diagonal
        if abs(c - col) == abs(r - row):
            return False

    return True

def NQ(n, row, tempArr, ansArr):
    if row == n:
        ansArr.append(tempArr.copy())
        return

    for col in range(n):
        if isSafe(row, col, tempArr):
            tempArr.append(col)
            NQ(n, row + 1, tempArr, ansArr)
            tempArr.pop()   # 🔥 backtrack

ansArr = []
n = 4
NQ(n, 0, [], ansArr)

# convert to board format
result = []
for sol in ansArr:
    board = []
    for col in sol:
        row_str = "." * col + "Q" + "." * (n - col - 1)
        board.append(row_str)
    result.append(board)

result
```

### 131. Palindrome Partitioning

```python
def PAL(s , curr , tempArr , ansArr) :
    if curr == len(s) :
        ansArr.append(tempArr.copy())
        return
    
    for i in range(curr , len(s)) :
        pal_str = s[curr:i+1]

        if pal_str == pal_str[::-1] :
            tempArr.append(pal_str)
            curr += len(pal_str)
            PAL(s , curr , tempArr , ansArr )
            curr -= len(pal_str)
            tempArr.pop()

ansArr = []
s = "abbccb"
PAL(s , 0 , [] , ansArr )

ansArr
```

### 2596. Check Knight Tour Configuration

```python
def knight( grid , row , col , total_box_go , n ) :
    if total_box_go == (n * n) :
        status = True
        print(True)
        return 

    c = 2
    for i in range(1,3) :
        if row + i < n and col + c < n :
            if grid[row + i][col + c] == total_box_go :
                knight(grid , row + i , col + c , total_box_go + 1 , n )
        if row + i < n and col - c >= 0 :
            if grid[row + i][col - c] == total_box_go :
                knight(grid , row + i , col - c , total_box_go + 1 , n )
        if row - i >= 0 and col + c < n :
            if grid[row - i][col + c] == total_box_go :
                knight(grid , row - i , col + c , total_box_go + 1 , n )
        if row - i >= 0 and col - c >= 0 :
            if grid[row - i][col - c] == total_box_go :
                knight(grid , row - i , col - c , total_box_go + 1 , n )
        
        c -= 1
        
    return False

status = False
grid = [[0,11,16,5,20],[17,4,19,10,15],[12,1,8,21,6],[3,18,23,14,9],[24,13,2,7,22]]
if grid[0][0] != 0 :
    print(False)
else :
    knight(grid , 0 , 0 , 1 , len(grid) )
```

### Leetcode : 90. Subsets II

```python

def findSub(nums , i , tempArr , ansArr) :
    if i == len(nums) :
        ansArr.append(tempArr.copy())
        return
    
    tempArr.append( nums[i] )
    findSub(nums , i + 1 , tempArr , ansArr)

    tempArr.pop()
    findSub(nums , i + 1 , tempArr , ansArr)

nums = [1,2,2]
ansArr = []
findSub(nums , 0 , [] , ansArr)

final_ans = []
for i in ansArr :   
    i.sort()
    if i not in final_ans :
        final_ans.append(i)

print(final_ans)
```

### Leetcode :  60. Permutation Sequence

```python
import math

class Solution:
    def getPermutation(self, n: int, k: int) -> str:

        nums = [str(i) for i in range(1, n + 1)]
        k -= 1                      # convert to 0-indexed

        ans = []

        while nums:
            fact = math.factorial(len(nums) - 1)

            idx = k // fact
            ans.append(nums.pop(idx))

            k %= fact

        return ''.join(ans)
```

### Leetcode : 46. Permutations

```python
class Solution:

    def findPer(self ,nums ,  le , tempArr , ansArr) :
        if len(tempArr) == le :
            if tempArr not in ansArr :
                ansArr.append(tempArr.copy())
            return
        
        for i in nums :
            if i in tempArr :
                continue
            tempArr.append(i)
            self.findPer(nums , le , tempArr , ansArr )
            tempArr.pop()
        

    def permute(self, nums):
        self.ans = []

        self.findPer(nums , len(nums) , [] , self.ans)

        return self.ans
```
