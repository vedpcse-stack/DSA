### Leetcode : 3. Longest Substring Without Repeating Characters

```python
arr = []
right = 0
left = 0
max_len = 0
s = "abcabcbb"
while right < len(s) :
    if s[right] not in arr :
        arr.append(s[right])
        right += 1
    else :
        max_len = max(max_len , len(arr))

        while s[right] in arr :
            arr.pop(0)
            left += 1

max(max_len , len(arr))
```
