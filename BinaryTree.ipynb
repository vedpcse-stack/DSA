### Leetcode : 94. Binary Tree Inorder Traversal

```python
def findIN(self , root , ans) :
    if root.left :
        self.findIN(root.left , ans)
    ans.append(root.val)
    if root.right :
        self.findIN(root.right , ans)
    return

def inorderTraversal(self, root):
    if not root :
        return []
        
    self.ans = []
    self.findIN(root , self.ans)

    return self.ans
```

### Leetcode : 144. Binary Tree Preorder Traversal

```python
def findPre(self , root , ans) :
    ans.append(root.val)

    if root.left :
        self.findPre(root.left , ans)
    if root.right :
        self.findPre(root.right , ans)
    return

def preorderTraversal(self, root):
    if not root :
        return []
        
    self.ans = []
    self.findPre(root , self.ans)

    return self.ans
```

### 145. Binary Tree Postorder Traversal

```python
def findPost(self , root , ans) :
    if root.left :
        self.findPost(root.left , ans)
    if root.right :
        self.findPost(root.right , ans)
    ans.append(root.val)
    return

def postorderTraversal(self, root):
    if not root :
        return []
        
    self.ans = []
    self.findPost(root , self.ans)

    return self.ans
```

### Leetcode : 199. Binary Tree Right Side View

```python
def findView(self , root , height , ansArr) :
    if height == len(ansArr):
        ansArr.append(root.val)
    else:
        ansArr[height] = root.val

    if root.left :
        self.findView(root.left , height+1 , ansArr)
    if root.right :
        self.findView(root.right , height+1 , ansArr)
    
    return

def rightSideView(self, root):
    if not root :
        return []

    self.ans = []
    self.findView(root , 0 , self.ans)

    return self.ans
```

### Leetcode : 987. Vertical Order Traversal of a Binary Tree

```python
def findOrder(self , root , row , col , ans) :
    if not ans.get((col)) :
        ans[col] = []
    
    ans[col].append((row , root.val))
    
    if root.left :
        self.findOrder(root.left , row + 1 , col - 1 , ans)
    if root.right :
        self.findOrder(root.right , row + 1 , col + 1 , ans)
    return

def verticalTraversal(self, root):
    if not root :
        return []

    d = {}
    self.findOrder(root , 0 , 0 , d)

    ans = []
    for i in sorted(d.keys()) :
        d[i].sort()
        ans.append([val for row , val in d[i]])
        

    return ans
```
