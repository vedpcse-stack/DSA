### Leetcode : 94. Binary Tree Inorder Traversal

```python
def findIN(self , root , ans) :
    if root.left :
        self.findIN(root.left , ans)
    ans.append(root.val)
    if root.right :
        self.findIN(root.right , ans)
    return

def inorderTraversal(self, root):
    if not root :
        return []
        
    self.ans = []
    self.findIN(root , self.ans)

    return self.ans
```

### Leetcode : 199. Binary Tree Right Side View

```python
def findView(self , root , height , ansArr) :
    if height == len(ansArr):
        ansArr.append(root.val)
    else:
        ansArr[height] = root.val

    if root.left :
        self.findView(root.left , height+1 , ansArr)
    if root.right :
        self.findView(root.right , height+1 , ansArr)
    
    return

def rightSideView(self, root):
    if not root :
        return []

    self.ans = []
    self.findView(root , 0 , self.ans)

    return self.ans
```

### Leetcode : 987. Vertical Order Traversal of a Binary Tree

```python
def findOrder(self , root , row , col , ans) :
    if not ans.get((col)) :
        ans[col] = []
    
    ans[col].append((row , root.val))
    
    if root.left :
        self.findOrder(root.left , row + 1 , col - 1 , ans)
    if root.right :
        self.findOrder(root.right , row + 1 , col + 1 , ans)
    return

def verticalTraversal(self, root):
    if not root :
        return []

    d = {}
    self.findOrder(root , 0 , 0 , d)

    ans = []
    for i in sorted(d.keys()) :
        d[i].sort()
        ans.append([val for row , val in d[i]])
        

    return ans
```

### Leetcode : 662. Maximum Width of Binary Tree

```python
from collections import deque
def widthOfBinaryTree(self, root):
    if not root:
        return 0

    q = deque([(root, 0)])   # (node, index)
    maxWidth = 0

    while q:
        level_size = len(q)

        # Leftmost index of this level
        first = q[0][1]

        # Rightmost index of this level
        last = q[-1][1]

        maxWidth = max(maxWidth, last - first + 1)

        for _ in range(level_size):
            node, index = q.popleft()

            # Normalize index to avoid very large numbers
            index -= first

            if node.left:
                q.append((node.left, 2 * index + 1))

            if node.right:
                q.append((node.right, 2 * index + 2))

    return maxWidth
```

### Leetcode : 110. Balanced Binary Tree

```python
def FindBalanced(self, root , ansArr ) :
    left = 0
    right = 0
    if root.left :
        left = self.FindBalanced(root.left , ansArr)
    if root.right :
        right = self.FindBalanced(root.right , ansArr)
    
    currNode = left - right
    if currNode in [-1 , 0 , 1] :
        ansArr.append(True)
    else :
        ansArr.append(False)
    
    return max(left , right) + 1

def isBalanced(self, root) -> bool:
    if not root :
        return True
    
    ans = []
    self.FindBalanced(root , ans)

    for i in ans :
        if not i :
            return False

    return True
```

### Leetcode : 543. Diameter of Binary Tree

```python
def findDai(self, root , ansArr ) :
    left = 0
    right = 0
    if root.left :
        left = self.findDai(root.left , ansArr)
    if root.right :
        right = self.findDai(root.right , ansArr)
    
    currNode = left + right
    ansArr.append(currNode)
    
    return max(left , right) + 1
def diameterOfBinaryTree(self, root) -> int:
    if not root :
        return True
    
    ans = []
    self.findDai(root , ans)

    return max(ans)
```

### Leetcode : 236. Lowest Common Ancestor of a Binary Tree

```python
def findAnc(self , root   , d , p , q , arr) :
    arr.append(root)
    if root == p or root == q :
        d[root] = arr.copy()
        
        
    if root.left :
        self.findAnc(root.left  , d , p , q , arr)
    
    if root.right :
        self.findAnc(root.right  , d , p , q , arr)
    
    arr.pop()
    
    return

def lowestCommonAncestor(self, root , p, q):
    if not root :
        return None

    d = defaultdict(list)
    arr = []
    self.findAnc(root , d , p , q , arr)

    if len(d[p]) >= len(d[q]) :
        small = d[q]
        large = d[p]
    else :
        small = d[p]
        large = d[q]

    for i in range(len(small)-1 , -1 , -1) :
        if small[i] in large :
            return small[i]
        
```

### Leetcode : 100. Same Tree

```python
def findIdn(self , p , q , ans) :
    if p and q :
        if p.val !=  q.val :
            ans.append(1)
    else :
        ans.append(1)
        return

    if p.left or q.left :
        self.findIdn(p.left , q.left , ans)     
    if p.right or q.right :
        self.findIdn(p.right , q.right , ans)  
    return        

def isSameTree(self, p, q):
    if not q and not p :
        return True
    if not q or not p :
        return False
    ans = []
    self.findIdn(p , q , ans)

    return sum(ans) == 0
```

### Leetcode : 103. Binary Tree Zigzag Level Order Traversal

```python
def findNodeByHeight(self , root , h , d) :
    d[h].append(root.val)

    if root.left :
        self.findNodeByHeight(root.left , h+1 , d)
    if root.right :
        self.findNodeByHeight(root.right , h+1 , d)
    
    return

def zigzagLevelOrder(self, root) :
    if not root :
        return []
    
    d = defaultdict(list)
    self.findNodeByHeight(root , 0 , d)

    ans = []
    for key , val in d.items() :
        if key % 2 == 0 :
            ans.append(val)
        else :
            ans.append(val[::-1])
    
    return ans
```

### Leetcode : 105. Construct Binary Tree from Preorder and Inorder Traversal

```python
def makeTree(self, preorder, inorder):
    if not preorder or not inorder:
        return None

    root = TreeNode(preorder[0])

    idx = inorder.index(preorder[0])

    left_inorder = inorder[:idx]
    right_inorder = inorder[idx + 1:]

    left_size = len(left_inorder)

    left_preorder = preorder[1:1 + left_size]
    right_preorder = preorder[1 + left_size:]

    root.left = self.makeTree(left_preorder, left_inorder)
    root.right = self.makeTree(right_preorder, right_inorder)

    return root

def buildTree(self, preorder, inorder) :
    return self.makeTree(preorder, inorder)
```

### Leetcode : 116. Populating Next Right Pointers in Each Node

```python
def getNodes(self , root , h , d) :
    d[h].append(root)

    if root.left :
        self.getNodes(root.left , h+1 , d)
    if root.right :
        self.getNodes(root.right , h+1 , d)
    
    return

def connect(self, root):
    if not root :
        return None

    d = defaultdict(list)
    self.getNodes(root , 0 , d)

    for i in d.values() :
        j = 0
        if len(i) == 1 :
            continue
            
        while j < len(i) - 2 :
            i[j].next = i[j+1]
            j += 1
        

        i[j].next = i[j+1]
    
    return d[0][0]
```

### Leetcode : 700. Search in a Binary Search Tree

```python
def searchBST(self, root, val: int):
    while root :
        if root.val == val :
            return root
        if root.val >= val :
            root = root.left
        else :
            root = root.right
    return None
```
