### Leetcode : 125. Valid Palindrome

```python
s = "A man, a plan, a canal: Panama"
new_s = ""

for i in s :
    if i.isalnum() :
        new_s = new_s + i.lower()

if new_s[ : :-1] == new_s :
    # return True
    print(True)
else :
    print(False)
```

### Leetcode : 242. Valid Anagram

```python
s = "anagram"
t = "nagaram"

if len(s) != len(t) :
    # return False
    print(False)

for i in s :
    if s.count(i) != t.count(i) :
        print(True)


print(False)
```

### Leetcode : 443. String Compression

```python
chars = ["a","a","b","b","c","c","c"]
s = ""
count = 1

for i in range(len(chars)) :
    
    if i < len(chars) - 1 and chars[i] == chars[i+1] :
        count = count + 1
        continue
    
    if count == 1 :
        s = s + chars[i] 
    else :
        s = s + chars[i] + str(count)
        
    count = 1

for i in range(len(s)) :
    chars[i] = s[i]

print(len(s))
```

### Leetcode : 49. Group Anagrams

```python
strs = ["eat","tea","tan","ate","nat","bat"]
out = {}

for word in strs :
    key = "".join(sorted(word))
    if key not in out :
        out[key] = []

    out[key].append(word)
list(out.values())
```

### Leetcode : 14. Longest Common Prefix

```python
strs = ["flower","flow","flight"]
s = ""

min_len = len(min(strs , key = len))

for i in range(min_len) :
    temp = min(strs , key = len)[i]
    for j in strs :
        if j[i] != temp :
            
            break
    else :
        s = s + temp
s
```

### Leetcode : 1392. Longest Happy Prefix

```python
out = ""
s = "ababab"

left = 0
right = len(s) - 1

while left < len(s) - 1 :
    pre = s[0 : left + 1]
    suff = s[right : ]

    if pre == suff :
        out = pre
    
    left += 1
    right -= 1

print(out)
```

### Leetcode : 1358. Number of Substrings Containing All Three Characters

```python
def numberOfSubstrings(self, s: str) -> int:
    left = 0
    count = 0
    freq = {'a': 0, 'b': 0, 'c': 0}

    for right in range(len(s)):
        freq[s[right]] += 1

        while freq['a'] > 0 and freq['b'] > 0 and freq['c'] > 0:
            count += len(s) - right
            freq[s[left]] -= 1
            left += 1

    return count
```
