### Array ( part 1 )
### leetcode :  121. Best Time to Buy and Sell Stock

```python
def maxProfit(prices):
    min_price = float('inf')
    max_profit = 0

    for price in prices:
        if price < min_price:
            min_price = price
        else:
            profit = price - min_price
            if profit > max_profit:
                max_profit = profit

    return max_profit

maxProfit( [7,1,4,5,6,3] )
```

### Array ( part 2 )
### LeetCode : 53. Maximum Subarray

```python
nums = [-2,1,-3,4,-1,2,1,-5,4]

max_sum = nums[0]
temp_sum = nums[0]

for i in nums[1:]:
    temp_sum = max(i, temp_sum + i)
    max_sum = max(max_sum, temp_sum)

print(max_sum)
```

### Array ( part 2 )
### Leetcode : 11. Container With Most Water

```python
# height = [1,8,6,2,5,4,8,3,7]
height = [1,1]

left_pointer = 0
right_pointer = len(height) - 1

max_water = 0

while left_pointer != right_pointer :

  min_line = min( height[left_pointer] , height[right_pointer] )

  temp_water = min_line*( (right_pointer - left_pointer) )
  max_water = max(max_water , temp_water)

  if height[left_pointer] < height[right_pointer] :
    left_pointer += 1
  else :
    right_pointer -= 1


print(max_water)
```

### Array ( part 2 )
### LeetCode : 75. Sort Colors

```python
nums = [2,0,2,1,1,0]
left = 0
mid = 0
right = len(nums) - 1

while mid <= right:

    if nums[mid] == 0:
        nums[left], nums[mid] = nums[mid], nums[left]
        left += 1
        mid += 1

    elif nums[mid] == 1:
        mid += 1

    else:  # nums[mid] == 2
        nums[mid], nums[right] = nums[right], nums[mid]
        right -= 1
```

### Array ( part 2)
### LeetCode : 15. 3Sum

```python
nums = [-1,0,1,2,-1,-4]

nums.sort()
ans = []

for i in range(len(nums)):
    if i > 0 and nums[i] == nums[i-1]:
        continue

    left = i + 1
    right = len(nums) - 1

    while left < right:
        total = nums[i] + nums[left] + nums[right]
        if total == 0:
            ans.append([nums[i], nums[left], nums[right]])
            left += 1
            right -= 1

            while left < right and nums[left] == nums[left-1]:
                left += 1

            while left < right and nums[right] == nums[right+1]:
                right -= 1

        elif total < 0:
            left += 1

        else:
            right -= 1

ans
```
