### Array ( part 1 )
### leetcode :  121. Best Time to Buy and Sell Stock

```python
def maxProfit(prices):
    min_price = float('inf')
    max_profit = 0

    for price in prices:
        if price < min_price:
            min_price = price
        else:
            profit = price - min_price
            if profit > max_profit:
                max_profit = profit

    return max_profit

maxProfit( [7,1,4,5,6,3] )
```

### LeetCode : 53. Maximum Subarray

```python
nums = [-2,1,-3,4,-1,2,1,-5,4]

max_sum = nums[0]
temp_sum = nums[0]

for i in nums[1:]:
    temp_sum = max(i, temp_sum + i)
    max_sum = max(max_sum, temp_sum)

print(max_sum)
```

### Array ( part 2 )
### Leetcode : 11. Container With Most Water

```python
# height = [1,8,6,2,5,4,8,3,7]
height = [1,1]

left_pointer = 0
right_pointer = len(height) - 1

max_water = 0

while left_pointer != right_pointer :

  min_line = min( height[left_pointer] , height[right_pointer] )

  temp_water = min_line*( (right_pointer - left_pointer) )
  max_water = max(max_water , temp_water)

  if height[left_pointer] < height[right_pointer] :
    left_pointer += 1
  else :
    right_pointer -= 1


print(max_water)
```

### LeetCode : 75. Sort Colors

```python
nums = [2,0,2,1,1,0]
left = 0
mid = 0
right = len(nums) - 1

while mid <= right:

    if nums[mid] == 0:
        nums[left], nums[mid] = nums[mid], nums[left]
        left += 1
        mid += 1

    elif nums[mid] == 1:
        mid += 1

    else:  # nums[mid] == 2
        nums[mid], nums[right] = nums[right], nums[mid]
        right -= 1
```

### Array ( part 2)
### LeetCode : 15. 3Sum

```python
nums = [-1,0,1,2,-1,-4]

nums.sort()
ans = []

for i in range(len(nums)):
    if i > 0 and nums[i] == nums[i-1]:
        continue

    left = i + 1
    right = len(nums) - 1

    while left < right:
        total = nums[i] + nums[left] + nums[right]
        if total == 0:
            ans.append([nums[i], nums[left], nums[right]])
            left += 1
            right -= 1

            while left < right and nums[left] == nums[left-1]:
                left += 1

            while left < right and nums[right] == nums[right+1]:
                right -= 1

        elif total < 0:
            left += 1

        else:
            right -= 1

ans
```

### Array ( part 2 )
### LeetCode : 18. 4Sum

```python
# nums = [-1,0,1,2,-1,-4]
# nums = [-2 , -1 , -1 , 1 , 1 , 2 , 2]
# nums = [1,0,-1,0,-2,2]
nums = [-3,-1,0,2,4,5]
target = 2
nums.sort()
ans = []

for k in range( len(nums) ) : # first pointer
    if k > 0 and nums[k] == nums[k-1]: # # not check for same combination of k
        continue
    
    for i in range(k+1 , len(nums)): # second poiner
        if i > k+1 and nums[i] == nums[i-1]: # not check for same combination of i
            continue
        
        left = i + 1 # pointer next to second pointer
        right = len(nums) - 1 # pointer to last value

        while left < right:
            total = nums[k] +  nums[i] + nums[left] + nums[right]
            if total == target:
                ans.append([nums[k] , nums[i], nums[left], nums[right]])
                left += 1
                right -= 1

                while left < right and nums[left] == nums[left-1]: # skip numbers from left till some unique number
                    left += 1

                while left < right and nums[right] == nums[right+1]: # skip numbers from right till some unique number
                    right -= 1

            elif total < target:
                left += 1

            else:
                right -= 1

ans
```

### Array ( part 2 )
### LeetCode : 74. Search a 2D Matrix

```python
matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]]
target = 13

# time compl : O( m ) --> not good we need still more good way
# for i in matrix :
#     if i.count(target) >= 1 :
#         print(True)
#         break
# else :
#     print(False)

def search_in_matrix(matrix , target) :
    row = len(matrix) 
    cols = len( matrix[0] )

    left = 0
    right = row - 1

    while left <= right :
        mid = left +  (right - left) // 2

        if matrix[mid][0] <= target <= matrix[mid][cols-1]  :
            return search_in_row(matrix[mid] , target)

        if target > matrix[mid][cols-1] :
            left = mid + 1
            continue
        else :
            right = mid - 1
    
    return False

def search_in_row(row , target) :
    m = len(row)
    left = 0
    right = m - 1

    while left <= right :
        mid = left + (right - left) // 2

        if row[mid] == target :
            return True
        
        if row[mid] > target :
            right = mid - 1
        else :
            left = mid + 1
    
    return False

print(search_in_matrix(matrix,target))
```
