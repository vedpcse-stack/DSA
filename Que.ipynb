### LeetCode : 74. Search a 2D Matrix

```python
matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]]
target = 13

# time compl : O( m ) --> not good we need still more good way
# for i in matrix :
#     if i.count(target) >= 1 :
#         print(True)
#         break
# else :
#     print(False)

def search_in_matrix(matrix , target) :
    row = len(matrix) 
    cols = len( matrix[0] )

    left = 0
    right = row - 1

    while left <= right :
        mid = left +  (right - left) // 2

        if matrix[mid][0] <= target <= matrix[mid][cols-1]  :
            return search_in_row(matrix[mid] , target)

        if target > matrix[mid][cols-1] :
            left = mid + 1
            continue
        else :
            right = mid - 1
    
    return False

def search_in_row(row , target) :
    m = len(row)
    left = 0
    right = m - 1

    while left <= right :
        mid = left + (right - left) // 2

        if row[mid] == target :
            return True
        
        if row[mid] > target :
            right = mid - 1
        else :
            left = mid + 1
    
    return False

print(search_in_matrix(matrix,target))
```

### Leetcode : 39. Combination Sum

```python
def findSum( candidates , target , i , tempArr , ansArr ) :
    
    if sum(tempArr) == target :
        # print(tempArr)
        ansArr.append(tempArr.copy())
        return
    
    if sum(tempArr) > target or i >= len(candidates) :
        return
    
    tempArr.append(candidates[i])
    findSum(candidates , target , i , tempArr , ansArr )
    findSum(candidates , target , i+1 , tempArr , ansArr )

    tempArr.pop()
    findSum(candidates , target , i+1 , tempArr , ansArr )

ansArr = []
target = 8
candidates = [2,3,5]
findSum(candidates , target , 0 , [] , ansArr)

final_ans = []

for i in ansArr :
    if final_ans.count(i) == 0 :
        final_ans.append(i)

final_ans
```

### Leetcode : 40. Combination Sum II

```python
def findSum(arr , target , i , tempArr , ansArr ) :
    if sum(tempArr) == target :
        print(tempArr)
        ansArr.append(tempArr.copy())
        return
    
    if sum(tempArr) > target or i >= len(arr)  :
        return
    
    tempArr.append(arr[i])
    findSum(arr , target , i+1 , tempArr , ansArr)

    tempArr.pop()

    # skip duplicates
    while i + 1 < len(arr) and arr[i] == arr[i+1]:
        i += 1

    findSum(arr , target , i+1 , tempArr , ansArr)

arr = [10,1,2,7,6,1,5]
target = 8
ansArr = []
arr.sort()

findSum(arr , target , 0 , [] , ansArr)

ansArr
```

### Recursion & Backtracking
### Leetcode : 51. N-Queens

```python
def isSafe( row, col, tempArr):
    for r in range(row):
        c = tempArr[r]

        # same column
        # if len(set(tempArr)) == row + 1:
        #     return False
        if c == col :
            return False 

        # diagonal
        if abs(c - col) == abs(r - row):
            return False

    return True

def NQ(n, row, tempArr, ansArr):
    if row == n:
        ansArr.append(tempArr.copy())
        return

    for col in range(n):
        if isSafe(row, col, tempArr):
            tempArr.append(col)
            NQ(n, row + 1, tempArr, ansArr)
            tempArr.pop()   # 🔥 backtrack

ansArr = []
n = 4
NQ(n, 0, [], ansArr)

# convert to board format
result = []
for sol in ansArr:
    board = []
    for col in sol:
        row_str = "." * col + "Q" + "." * (n - col - 1)
        board.append(row_str)
    result.append(board)

result
```

### 131. Palindrome Partitioning

```python
def PAL(s , curr , tempArr , ansArr) :
    if curr == len(s) :
        ansArr.append(tempArr.copy())
        return
    
    for i in range(curr , len(s)) :
        pal_str = s[curr:i+1]

        if pal_str == pal_str[::-1] :
            tempArr.append(pal_str)
            curr += len(pal_str)
            PAL(s , curr , tempArr , ansArr )
            curr -= len(pal_str)
            tempArr.pop()

ansArr = []
s = "abbccb"
PAL(s , 0 , [] , ansArr )

ansArr
```

```python
def findSub(nums , i , tempArr , ansArr) :
    if i == len(nums) :
        ansArr.append(tempArr.copy())
        return
    
    tempArr.append( nums[i] )
    findSub(nums , i + 1 , tempArr , ansArr)

    tempArr.pop()
    findSub(nums , i + 1 , tempArr , ansArr)

nums = [1,2,2]
ansArr = []
findSub(nums , 0 , [] , ansArr)

final_ans = []
for i in ansArr :   
    i.sort()
    if i not in final_ans :
        final_ans.append(i)

print(final_ans)
```

#### Leetcode : 142. Linked List Cycle II

```python
def detectCycle(self, head) :
    slow = head
    fast = head
    IsCycle = False

    while fast and fast.next :
        slow = slow.next
        fast = fast.next.next

        if slow == fast :
            IsCycle = True
            break
    
    if not IsCycle :
        return None
    
    slow = head
    while slow != fast :
        slow = slow.next
        fast = fast.next
    
    return slow
```

```python
def hasCycle(self, head):
    slow = fast = head

    while fast and fast.next :
        slow = slow.next
        fast = fast.next.next

        if slow == fast :
            return True
    
    return False
```

#### Linked List (part 1)
#### Leetcode : 21. Merge Two Sorted Lists

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeTwoLists(self, list1 , list2) :
        dummy = ListNode()
        cur = dummy

        while list1 and list2:
            if list1.val > list2.val:
                cur.next = list2
                list2 = list2.next
            else:
                cur.next = list1
                list1 = list1.next
            cur = cur.next

        if list1:
            cur.next = list1
        else:
            cur.next = list2

        return dummy.next
```

#### Leetcode : 234. Palindrome Linked List

```python
def isPalindrome(self, head) :
    slow = fast = head
    stack = []
    while fast and fast.next:
        stack.append(slow.val)
        slow = slow.next
        fast = fast.next.next
    if fast:
        slow = slow.next
    while slow:
        if stack.pop() != slow.val:
            return False
        slow = slow.next
    
    return True
```
